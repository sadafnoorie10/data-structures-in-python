    occurrences = []
    start = 0
    
    # Keep searching for the substring and updating the starting index
    while True:
        start = main_str.find(sub_str, start)

        occurrences.append(start)
        start += 1  # Move past the current match to avoid overlapping
        
    return occurrences

Q6) write a code to perform basic string compression using the counts of repeated characters.

A6) Here's a Python code that performs basic string compression by counting repeated characters:

```python
# Function to perform basic string compression
def compress_string(s):
    compressed = []
    count = 1
    
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            compressed.append(s[i - 1] + str(count))
            count = 1
    
    # Append the last character and its count
    compressed.append(s[-1] + str(count))
    
    # Join the compressed list to form the final string
    compressed_string = ''.join(compressed)
    
    # Return the original string if the compressed string is not shorter
    return compressed_string if len(compressed_string) < len(s) else s

# Example usage
input_string = "aaabbccccdaa"
compressed_string = compress_string(input_string)
print("Compressed String:", compressed_string)
```

### Explanation:
- The code compresses the string by counting consecutive repeated characters.
- If the compressed version is shorter than the original string, it returns the compressed string; otherwise, it returns the original string.
  
For example, the input `"aaabbccccdaa"` would be compressed to `"a3b2c4d1a2"`. If the compression does not reduce the size, it will return the original string.

    first, second = float('-inf'), float('-inf')  # Initialize to negative infinity

            first = number   # Update first largest
        elif first > number > second:
            second = number   # Update second largest if number is between first and second

    return second if second != float('-inf') else None  # Return None if no second largest found

# Example usage

### Explanation:

   - If a number is greater than `first`, it updates `second` to be `first` and then updates `first` to the current number.
   - If a number is between `first` and `second`, it updates `second` to that number.
3. Finally, it returns `second`, or `None` if no valid second largest number is found.

Q14) write a code to reverse a list in-place without using any built-in reverse functions.

A14) Here’s a Python code to reverse a list in-place without using any built-in reverse functions:

```python
# Function to reverse a list in-place
def reverse_list(lst):
    left = 0  # Start index
    right = len(lst) - 1  # End index

    while left < right:
        # Swap the elements at left and right indices
        lst[left], lst[right] = lst[right], lst[left]
        
        # Move the indices towards the center
        left += 1
        right -= 1

# Example usage
input_list = [1, 2, 3, 4, 5]
print("Original List:", input_list)

reverse_list(input_list)

print("Reversed List:", input_list)
```

### Explanation:
1. The function initializes two pointers, `left` and `right`, at the start and end of the list, respectively.
2. While the `left` index is less than the `right` index:
   - It swaps the elements at the `left` and `right` indices.
   - It then moves the `left` index one step to the right and the `right` index one step to the left.
3. The process continues until all elements have been swapped.

For example, given the input list `[1, 2, 3, 4, 5]`, the output will be `[5, 4, 3, 2, 1]`.

   - If the current element is less than the previous element, it sets `ascending` to `False`.

```python
# Function to merge two sorted lists into a single sorted list
def merge_sorted_lists(list1, list2):
    merged_list = []
    i, j = 0, 0  # Pointers for each list

    # Compare elements from both lists and merge them in sorted order
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1

    # Add any remaining elements from list1
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1

    # Add any remaining elements from list2
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1

    return merged_list

# Example usage
sorted_list1 = [1, 3, 5, 7]
sorted_list2 = [2, 4, 6, 8]
result = merge_sorted_lists(sorted_list1, sorted_list2)

print("Merged Sorted List:", result)
```

### Explanation:
1. The function initializes an empty list `merged_list` to store the merged result.
2. Two pointers, `i` and `j`, are used to track the current index of `list1` and `list2`, respectively.
3. The first `while` loop compares elements from both lists and appends the smaller element to `merged_list`, advancing the corresponding pointer.
4. After one of the lists is fully traversed, the next two `while` loops append any remaining elements from `list1` and `list2` to `merged_list`.
5. Finally, the function returns the merged sorted list.

### Example Output:
For the example input lists `[1, 3, 5, 7]` and `[2, 4, 6, 8]`, the output will be `[1, 2, 3, 4, 5, 6, 7, 8]`.

def shuffle_list(lst):

        lst[i], lst[j] = lst[j], lst[i]
